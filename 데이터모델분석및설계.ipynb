```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
data = pd.read_csv('/content/drive/MyDrive/A회사_보습제_매출데이터_v1.csv', encoding='cp949')
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
months = ['Sep','Oct','Nov','Dec','Jan' , 'Fev']
```

```python
total_sales = df.iloc[0][1:]
target_sales = df.iloc[1][1:]
advertising_cost = df.iloc[2][1:]
social_network_cost  = df.iloc[3][1:]
price_per_ounce  = df.iloc[4][1:]
```

```python
plt.plot(months, total_sales, label='Total Sales')
plt.plot(months,target_sales,label='Target Sales')

plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Target Sales vs Total Sales')  # 그래프 제목
plt.legend()  # 범례 표시
plt.show()  # 그래프 보여주기
```

```python
plt.figure(figsize=(20, 2))
plt.subplot(1, 2, 1)
plt.plot(months, advertising_cost,color='r', label='Advertising Cost')
plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Advertising Cost')  # 그래프 제목
plt.gca().invert_yaxis()
plt.legend()  # 범례 표시


plt.subplot(1, 2, 2)
plt.plot(months, total_sales, label='Total Sales')
plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Total Sales')  # 그래프 제목
plt.legend()  # 범례 표시
plt.show()  # 그래프 보여주기
```

```python
plt.figure(figsize=(20, 2))
plt.subplot(1, 2, 1)
plt.plot(months,social_network_cost, color='m', label='Social Network Cost')
plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Social Network Cost')  # 그래프 제목
plt.legend()  # 범례 표시

plt.subplot(1, 2, 2)
plt.plot(months, total_sales, label='Total Sales')
plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Total Sales')  # 그래프 제목
plt.legend()  # 범례 표시
plt.show()  # 그래프 보여주기
```

```python
plt.plot(months,price_per_ounce,label='price_per_ounce', marker='o', linestyle='None')
plt.gca().invert_yaxis()
plt.plot(months, total_sales, label='Total Sales')
plt.gca().invert_yaxis()
plt.xlabel('Month')  # x축 라벨
plt.ylabel('Cost')  # y축 라벨
plt.title('Price Per Ounce vs Total Sales')  # 그래프 제목
plt.legend()  # 범례 표시
plt.show()  # 그래프 보여주기
```

```python
df = df.transpose()
df.columns = ['Total Sales', 'Target Sales', 'Advertising Cost','Social Network Cost', 'Price Per Ounce']

df = df.drop('Unnamed: 0', axis=0)
```

```python
df
```

```python
for column in df.columns:
    df[column] = df[column].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data = df.corr(), annot=True,
fmt = '.2f', linewidths=.5, cmap='Blues')
```
